```python
names=iris["species"]
iris['species']=iris['species'].map({'Iris-versicolor':0,'Iris-setosa':1,'Iris-virginica':2})
iris['species'].unique()
```

## Feature engineering 

```python
features=iris .drop("species",axis=1)
target=iris['species']

# rarget will be our dependent variable 

# Example: Display features and target shapes (optional)
print("Features shape:", features.shape)
print("Target shape:", target.shape)
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris


# Step 4: Split data into training and testing sets
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Step 5: Define the parameters for GridSearchCV
parameters = {
    "criterion": ['gini', 'entropy'],
    "max_depth": [1, 2, 3, 4, 5],
    "min_samples_split": [3, 5, 7, 9],
    "max_features": ["sqrt", "log2"],
    "n_estimators": [20, 40, 60, 80, 100]
}
```

```python

# Step 6: Initialize the RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model
model.fit(features_train, target_train)

# Step 7: Predict on the testing data
target_pred = model.predict(features_test)

# Step 8: Evaluate the model's accuracy
accuracy = accuracy_score(target_test, target_pred)
print(f"Accuracy : {accuracy}")
```

```python
import pandas as pd

label_to_species = {0: 'Iris-versicolor', 1: 'Iris-setosa', 2: 'Iris-virginica'}

def predict_flower(sepal_length, sepal_width, petal_length, petal_width):
    input_features = pd.DataFrame({
        'sepal_length': [sepal_length],
        'sepal_width': [sepal_width],
        'petal_length': [petal_length],
        'petal_width': [petal_width]
    })
    print("\n\nInput Features: ", input_features)

    
    predicted_species_num = model.predict(input_features)[0]
    predicted_species_name = label_to_species[predicted_species_num]
    
    return predicted_species_name

predicted_flower = predict_flower(4.0, 2.6, 0.4, 0.2)
print(f"The predicted flower type is: {predicted_flower}")

predicted_flower2 = predict_flower(5.0, 4.6, 5.4, 1.2)
print(f"The predicted flower type is: {predicted_flower2}")
```

```python
from sklearn.metrics import classification_report, accuracy_score
report = classification_report(target_test, target_pred, target_names=label_to_species.values())
print("\nClassification Report:")
print(report)
```
